# TP4 - Watermark Detection

```python
import numpy as np
import copy
import cv2
from matplotlib import pyplot as plt
```

## I - Watermark embedding and Channel modeling

```python
# parameters
magn = 1 # magnitude of the WM
theta = 0.5 # density of the WM
sigma = 1 # std of the attack
```

```python
# Reads in the image as floats but with uint8 values (min= 0.0, max=255.0)
x = cv2.imread('liftingbody.png', cv2.IMREAD_GRAYSCALE).astype('float')
print(x.shape)
print(x.max())
plt.imshow(x)
plt.title('x')
plt.axis('off')
```

```python
wp = (2 * magn * np.random.randint(2, size=x.shape) - magn).astype(float)
print(wp.dtype)
print(wp[:5,:5])
plt.imshow(wp)
plt.title("w'")
plt.axis('off')
```

```python
# number of changed pixels
N_change = int(np.prod(wp.shape)*theta)
# density mask for the WM
mask = [1]*N_change + [0]*(np.prod(wp.shape) - N_change)
np.random.shuffle(mask)
mask = np.reshape(mask, wp.shape)
w = wp * mask

# verify the distribution
print(w.dtype)
plt.hist(w.flatten(),3)
plt.title('w dist')
```

```python
# WM the image
y = x+w
# Random attack
z = sigma * np.random.randn(*(y.shape))
print(z.shape)
print(z.dtype)
plt.hist(z.flatten())
plt.title('z dist')
v = y+z
```

```python
to_show = {'original':x,
           'watermarked':y,
           'wm_attacked':v}
plt.figure(figsize=(20, 15))
for i, im in enumerate(to_show):
    plt.subplot(1,len(to_show),i+1)
    plt.imshow(to_show[im])
    plt.axis('off')
    plt.title(im)
```

There is almost no visible degradation of the original image, the WM is impossible to be detected visualy and the attack may cause some very slight modifications but they are hard to spot.

## 2 - Non blind detection

```python
def wm_estim_nb(v, x):
    """Returns the estimated WM for non blind detection
    v: image received
    x: original image
    """
    return v-x
```

```python
def correlation(wh, w):
    """Returns the WM correlation
    wh: estimated WM
    w: original WM
    """
    return np.dot(wh.flatten(),w.flatten())/np.sum(w!=0)
```

```python
def add_wm(x, magn, theta):
    """Returns the watermarked image
    x: original image
    magn: magnitude of the WM
    theta: density of the WM
    """
    wp = (2 * magn * np.random.randint(2, size=x.shape) - magn).astype(float)
    N_change = int(np.prod(wp.shape)*theta)
    mask = [1]*N_change + [0]*(np.prod(wp.shape) - N_change)
    np.random.shuffle(mask)
    mask = np.reshape(mask, wp.shape)
    w = wp * mask
    return x+w, w

def add_attack(y, sigma):
    """Returns the attacked image
    y: input image
    sigma: std of the attack
    """
    z = sigma * np.random.randn(*(y.shape))
    return y+z

# number of tries
n_try = 50
results_nb = []
for i in range(n_try):
    y, w = add_wm(x, magn, theta)
    v = add_attack(y, sigma)
    wh = wm_estim_nb(v, x)
    results_nb.append(correlation(wh, w))

plt.plot(np.arange(n_try), results_nb)
plt.plot(np.arange(n_try), [np.mean(results_nb)]*n_try, '--')
plt.legend()
plt.title('Non-blind correlation\nrho_mean=%.3f'%np.mean(results_nb))
```

## 3 - Blind detection

```python
from skimage.util.shape import view_as_windows

def wm_estim_b(v, win_size=3):
    """Returns the estimated WM for blind detection
    v: image received
    win_size: the window size for the local averaging
    """
    assert win_size>0, "the window size for the local averaging should be > 0"
    # pad original data in a symmetrical manner to have same output size as v
    xh = np.concatenate([v, v[:,-win_size+1:][:,::-1]], axis=1)
    xh = np.concatenate([xh, xh[-win_size+1:][::-1]], axis = 0)
    xh = view_as_windows(xh, (win_size, win_size))
    # local means
    xh = np.mean(xh.reshape([*v.shape, -1]), axis=-1)
    return v-xh
```

```python
# number of tries
n_try = 50
results_b = []
for i in range(n_try):
    y, w = add_wm(x, magn, theta)
    v = add_attack(y, sigma)
    wh = wm_estim_b(v)
    results_b.append(correlation(wh, w))

plt.plot(np.arange(n_try), results_b)
plt.plot(np.arange(n_try), [np.mean(results_b)]*n_try, '--')
plt.title('Blind correlation\nrho_mean=%.3f'%np.mean(results_b))
```

```python
plt.plot(np.arange(n_try), results_nb, label='Non-blind')
plt.plot(np.arange(n_try), [np.mean(results_nb)]*n_try, '--', label='AVG NB=%.3f'%np.mean(results_nb))
plt.plot(np.arange(n_try), results_b, label='Blind')
plt.plot(np.arange(n_try), [np.mean(results_b)]*n_try, '--', label='AVG B=%.3f'%np.mean(results_b))
plt.legend()
```

Because of the unexact estimation of the WM in the blind setup, in the presence of the WM, the linear correlation with the original WM is less high and varies more than the non-blind detection but still good.
