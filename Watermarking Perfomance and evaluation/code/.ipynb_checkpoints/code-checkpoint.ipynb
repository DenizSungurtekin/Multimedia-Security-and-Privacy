```python
import numpy as np
import copy
import cv2
from matplotlib import pyplot as plt
```

## I - Watermark embedding and Channel modeling

```python
# parameters
magn = 1 # magnitude of the WM
theta = 0.5 # density of the WM
sigma = 1 # std of the attack
```

```python
# Reads in the image as floats but with uint8 values (min= 0.0, max=255.0)
x = cv2.imread('cameraman.tif', cv2.IMREAD_GRAYSCALE).astype('float')
print(x.shape)
print(x.max())
plt.imshow(x)
plt.title('x')
plt.axis('off')
```

```python
wp = (2 * magn * np.random.randint(2, size=x.shape) - magn).astype(float)
print(wp.dtype)
print(wp[:5,:5])
plt.imshow(wp)
plt.title("w'")
plt.axis('off')
```

```python
# number of changed pixels
N_change = int(np.prod(wp.shape)*theta)
# density mask for the WM
mask = [1]*N_change + [0]*(np.prod(wp.shape) - N_change)
np.random.shuffle(mask)
mask = np.reshape(mask, wp.shape)
w = wp * mask

# verify the distribution
print(w.dtype)
plt.hist(w.flatten(),3)
plt.title('w dist')
```

```python
# WM the image
y = x+w
# Random attack
z = sigma * np.random.randn(*(y.shape))
print(z.shape)
print(z.dtype)
plt.hist(z.flatten())
plt.title('z dist')
v = y+z
```

```python
to_show = {'original':x,
           'watermarked':y,
           'wm_attacked':v}
plt.figure(figsize=(20, 15))
for i, im in enumerate(to_show):
    plt.subplot(1,len(to_show),i+1)
    plt.imshow(to_show[im])
    plt.axis('off')
    plt.title(im)
```

There is almost no visible degradation of the original image, the WM is impossible to be detected visualy and the attack may cause some very slight modifications but they are hard to spot.

## 2 - Non blind detection

```python
def wm_estim_nb(v, x):
    """Returns the estimated WM for non blind detection
    v: image received
    x: original image
    """
    return v-x
```

```python
def correlation(wh, w,N):
    """Returns the WM correlation
    wh: estimated WM
    w: original WM
    """
    return np.sum(np.dot(wh.flatten(),w.flatten()))/N
```

```python
def add_wm(x, magn, theta):
    """Returns the watermarked image
    x: original image
    magn: magnitude of the WM
    theta: density of the WM
    """
    wp = (2 * magn * np.random.randint(2, size=x.shape) - magn).astype(float)
    N_change = int(np.prod(wp.shape)*theta)
    mask = [1]*N_change + [0]*(np.prod(wp.shape) - N_change)
    np.random.shuffle(mask)
    mask = np.reshape(mask, wp.shape)
    w = wp * mask
    return x+w, w, N_change

def add_attack(y, sigma):
    """Returns the attacked image
    y: input image
    sigma: std of the attack
    """
    z = sigma * np.random.randn(*(y.shape))
    return y+z

```

```python
def computeMean_Var(sigma,theta,magn):

  n_try = 100
  results_nb_H0 = []
  results_nb_H1 = []

  for i in range(n_try):

      # v = x + z 
      y, w , N = add_wm(x, magn, theta)
      vH1 = add_attack(y, sigma)

      # v = x + z 
      vH0 = add_attack(x, sigma)


      wh_H0 = wm_estim_nb(vH0, x) # z
      wh_H1 = wm_estim_nb(vH1, x) # z + w

      results_nb_H0.append(correlation(wh_H0, w, N))
      results_nb_H1.append(correlation(wh_H1, w, N))

  plt.figure()
  plt.plot(np.arange(n_try), results_nb_H0)
  plt.plot(np.arange(n_try), [np.mean(results_nb_H0)]*n_try, 'r-',label="Mean H0")
  plt.plot(np.arange(n_try), [np.var(results_nb_H0)]*n_try, 'b-',label="Variance H0")
  plt.legend()
  plt.title("rho_mean={mean} \nrho_variance={variance}".format(mean=np.mean(results_nb_H0), variance=np.var(results_nb_H0)))


  plt.figure()
  plt.plot(np.arange(n_try), results_nb_H1)
  plt.plot(np.arange(n_try), [np.mean(results_nb_H1)]*n_try, 'r-',label="Mean H1")
  plt.plot(np.arange(n_try), [np.var(results_nb_H1)]*n_try, 'b-',label="Variance H1")
  plt.legend()
  plt.title("rho_mean={mean} \nrho_variance={variance}".format(mean=np.mean(results_nb_H1), variance=np.var(results_nb_H1)))

  Mean_H0 = np.mean(results_nb_H0)
  Mean_H1 = np.mean(results_nb_H1)

  Var_H0 = np.var(results_nb_H0)
  Var_H1 = np.var(results_nb_H1)

  return Mean_H0,Var_H0,Mean_H1,Var_H1
```

```python
sigmas = [100]
thetas = [0.3]
magns = [5]

MeanH0 = []
VarH0 = []

MeanH1 = []
VarH1 = []

for sigma in sigmas:
  for theta in thetas:
    for magn in magns:
       m1,s1,m2,s2 = computeMean_Var(sigma,theta,magn)
       MeanH0.append(m1)
       VarH0.append(s1)
       MeanH1.append(m2)
       VarH1.append(s2)
        
```

```python
# pm,pf,pd detection

def plotcurves(sigma,theta,magn):

  n_try = 100
  results_nb_H0 = []
  results_nb_H1 = []

  for i in range(n_try):

    # v = x + z 
    y, w,N = add_wm(x, magn, theta)
    vH1 = add_attack(y, sigma)
 
    # v = x + z 
    vH0 = add_attack(x, sigma)


    wh_H0 = wm_estim_nb(vH0, x) # z
  
    wh_H1 = wm_estim_nb(vH1, x) # z + w

    results_nb_H0.append(correlation(wh_H0, w,N))
    results_nb_H1.append(correlation(wh_H1, w,N))


  total = results_nb_H0 + results_nb_H1

  T_rho = np.linspace(min(total),max(total),200)

  pd = []
  pf = []
  pm = []
  # classfication
  for t in T_rho:
    pm_c = 0
    pf_c = 0
    for h0 in results_nb_H0:
      if h0 > t:
          pm_c += 1
    
    for h1 in results_nb_H1:
      if h1 < t:
        pf_c += 1

    pm.append(pf_c/n_try)
    pf.append(pm_c/n_try)
    pd.append(1 - pf_c/n_try)  

  plt.figure()
  plt.plot(T_rho,pm,label="$p_m$")
  plt.plot(T_rho,pf,label="$p_f$")
  plt.plot(T_rho,pd,label="$p_d$")
  plt.xlabel(r"$T_\rho$")
  plt.ylabel("p")
  plt.legend()
  plt.grid()
  plt.title("$\sigma$ : "+str(sigma)+r"  $\theta_N$ : "+str(theta)+r"  $\gamma : \pm$"+str(magn))
  plt.show()

  plt.figure()
  plt.plot(pf,pm,label="ROC")
  plt.xlabel("$p_f$")
  plt.ylabel("$p_m$")
  plt.legend()
  plt.grid()
  plt.title("$\sigma$ : "+str(sigma)+r"  $\theta_N$ : "+str(theta)+r"  $\gamma : \pm$"+str(magn))
  plt.show()        
```

```python
# Plot Curves pm,pf,pd and ROC

# parameters
sigmas = [50,100]
thetas = [0.1,0.3]
magns = [1,5]


for sigma in sigmas:
  for theta in thetas:
    for magn in magns:
      plotcurves(sigma,theta,magn)

```

## 3 - Blind detection

```python
from skimage.util.shape import view_as_windows

def wm_estim_b(v, win_size=3):
    """Returns the estimated WM for blind detection
    v: image received
    win_size: the window size for the local averaging
    """
    assert win_size>0, "the window size for the local averaging should be > 0"
    # pad original data in a symmetrical manner to have same output size as v
    xh = np.concatenate([v, v[:,-win_size+1:][:,::-1]], axis=1)
    xh = np.concatenate([xh, xh[-win_size+1:][::-1]], axis = 0)
    xh = view_as_windows(xh, (win_size, win_size))
    # local means
    xh = np.mean(xh.reshape([*v.shape, -1]), axis=-1)
    return v-xh
```

```python
# number of tries
n_try = 50
results_b = []
for i in range(n_try):
    y, w = add_wm(x, magn, theta)
    v = add_attack(y, sigma)
    wh = wm_estim_b(v)
    results_b.append(correlation(wh, w))

plt.plot(np.arange(n_try), results_b)
plt.plot(np.arange(n_try), [np.mean(results_b)]*n_try, '--')
plt.title('Blind correlation\nrho_mean=%.3f'%np.mean(results_b))
```
